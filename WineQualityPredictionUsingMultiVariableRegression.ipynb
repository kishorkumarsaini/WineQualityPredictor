```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.model_selection import train_test_split
%matplotlib inline

dataset=pd.read_csv('winequality.csv')
dataset.isnull().any()
dataset.fillna(method='ffill')

x = dataset[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates','alcohol']].values
y = dataset['quality'].values

#lets check the average value of quality column
plt.figure(figsize=(15,10))
plt.tight_layout()
sns.distplot(dataset['quality'])

#split the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)

#train the model
regression=LinearRegression()
regression.fit(x_train,y_train)

#predict the data
y_predict=regression.predict(x_test)

df=pd.DataFrame({'Actual':y_test,'Predicted':y_predict})

df1=df.head(25)
df1.plot(kind='bar',figsize=(15,10))
plt.show()
#calculate the error

print("Mean absolute error",metrics.mean_absolute_error(y_test,y_predict))
print("Mean square error",metrics.mean_squared_error(y_test,y_predict))
print("Root Mean square error",np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
```
